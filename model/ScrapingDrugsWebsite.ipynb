```python
# URL of the website to scrape
url1="https://www.drugs.com"
response = requests.get(url1)
# Parse the HTML content using BeautifulSoup
soup = BeautifulSoup(response.content, "html.parser")
l=[]
links=[]
for row in soup.select('nav.ddc-paging a')[:27]:
 l.append(url1+row['href'])
for url in l :

# Send an HTTP request to the URL of the webpage you want to access
 response = requests.get(url)
# Parse the HTML content using BeautifulSoup
 soup = BeautifulSoup(response.content, "html.parser")

# Extract the text content of the webpage

 for row1 in soup.select('ul.ddc-list-column-2 a'):
  links.append(url1+row1['href'])
print(links)
```

```python
l=[]
for url in links:
 response = requests.get(url)
# Parse the HTML content using BeautifulSoup

 soup = BeautifulSoup(response.content, "html.parser")
 description=''
 for row in soup.select('p'):
  description+=row.get_text()
  description+='\n'
 print(description)

 p_tag = soup.find('h1')
 name1=p_tag.get_text()
 l.append([description,name1])
df = pd.DataFrame(l, columns=['description', 'name'])
df
```

```python
df.to_csv('drugs.csv', sep=',', index=False, encoding='utf-8')
```
